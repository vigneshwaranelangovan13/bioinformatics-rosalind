````python
# Rosalind Problem #5 - Dictionaries

# Step 1: Read the text from practice_5.txt
with open("practice_5.txt", "r") as f:
    text = f.read().strip()

# Step 2: Split text into words
words = text.split()

# Step 3: Create a dictionary to count word frequencies
word_counts = {}

for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# Step 4: Print results
for word, count in word_counts.items():
    print(word, count)
````

####**TASK BP21.2** <font color=orange> **Complete Rosalind #8  Translating RNA into Protein**</font>

````python
# Rosalind Problem #8 - Translating RNA into Protein

# Step 1: Define the RNA string directly
rna = "AUGGCCAUGGCGCCCAGAACUGAGAUCAAUAGUACCCGUAUUAACGGGUGA"

# Step 2: Codon translation table
codon_table = {
    "UUU":"F","CUU":"L","AUU":"I","GUU":"V",
    "UUC":"F","CUC":"L","AUC":"I","GUC":"V",
    "UUA":"L","CUA":"L","AUA":"I","GUA":"V",
    "UUG":"L","CUG":"L","AUG":"M","GUG":"V",
    "UCU":"S","CCU":"P","ACU":"T","GCU":"A",
    "UCC":"S","CCC":"P","ACC":"T","GCC":"A",
    "UCA":"S","CCA":"P","ACA":"T","GCA":"A",
    "UCG":"S","CCG":"P","ACG":"T","GCG":"A",
    "UAU":"Y","CAU":"H","AAU":"N","GAU":"D",
    "UAC":"Y","CAC":"H","AAC":"N","GAC":"D",
    "UAA":"Stop","CAA":"Q","AAA":"K","GAA":"E",
    "UAG":"Stop","CAG":"Q","AAG":"K","GAG":"E",
    "UGU":"C","CGU":"R","AGU":"S","GGU":"G",
    "UGC":"C","CGC":"R","AGC":"S","GGC":"G",
    "UGA":"Stop","CGA":"R","AGA":"R","GGA":"G",
    "UGG":"W","CGG":"R","AGG":"R","GGG":"G"
}

# Step 3: Translate RNA → Protein
protein = ""
for i in range(0, len(rna), 3):
    codon = rna[i:i+3]
    if codon in codon_table:
        aa = codon_table[codon]
        if aa == "Stop":
            break
        protein += aa

# Step 4: Print result
print(protein)

````

#<font color=blue>**P22.**</font> SeqIO – a function to parse( ) sequence files

####**TASK BP 22.1** Read in the practice dataset from Rosalind #9 and print the id of each sequence in the file.

````python
# Rosalind Problem #9 extension – GC content for each sequence

# Step 1: Read the FASTA file
with open("practice_9.fasta", "r") as f:
    fasta_data = f.read().strip().split(">")[1:]  # split into records

# Step 2: Process each record
for record in fasta_data:
    lines = record.splitlines()
    seq_id = lines[0]        # first line is the ID
    sequence = "".join(lines[1:])  # rest is the sequence

    # Step 3: Compute GC content
    gc_count = sequence.count("G") + sequence.count("C")
    gc_content = (gc_count / len(sequence)) * 100

    # Step 4: Print result
    print(seq_id, f"{gc_content:.2f}%")

````

#<font color=blue>**P23.**</font> SeqUtils.gc_fraction( ) – a function to compute GC content

####**TASK P23.1** Modify the script from TASK P22.1 so that it prints the GC content for each sequence in the file

####**TASK P23.2** <font color=orange> **Complete Rosalind #9  Computing GC Content**</font>

````python
# Rosalind Problem #9 - Computing GC Content

# Step 1: Read FASTA file
with open("practice_9.fasta", "r") as f:
    fasta_data = f.read().strip().split(">")[1:]  # split by '>' and skip first empty

# Step 2: Store GC content in a dictionary
gc_dict = {}

for record in fasta_data:
    lines = record.splitlines()
    seq_id = lines[0]                   # first line is ID
    sequence = "".join(lines[1:])       # join sequence lines
    gc_count = sequence.count("G") + sequence.count("C")
    gc_content = (gc_count / len(sequence)) * 100
    gc_dict[seq_id] = gc_content

# Step 3: Find sequence with highest GC content
max_id = max(gc_dict, key=gc_dict.get)
max_gc = gc_dict[max_id]

# Step 4: Print result (Rosalind output format)
print(max_id)
print(f"{max_gc:.6f}")
````
